```python
import pandas as pd

df = pd.read_csv("data_jobs.csv")

print(df)
df
```

```python
job_counts = df.job_title_short.value_counts().head(5)
```

```python
import matplotlib.pyplot as plt
plt.bar(job_counts.index, job_counts)
```

```python
job_counts = df.job_title_short.value_counts()
```

```python
#for a horizontal bar use .barh
plt.barh(job_counts.index, job_counts)
```

```python
#to change the horizontal bar to ascending order
job_counts = job_counts.sort_values(ascending = True) 
plt.barh(job_counts.index, job_counts)
```

```python
#matplotlib labelling. to clean up visualization 
# plt.xticks(rotation = 45, ha= 'right')- helps in aligning the data

job_counts = job_counts.sort_values(ascending= False) 
plt.bar(job_counts.index, job_counts)
plt.title('Postings by Job Title')
plt.ylabel('count of job postings')
plt.xticks(rotation = 45, ha= 'right')
plt.show()
```
